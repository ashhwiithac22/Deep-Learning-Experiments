```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.preprocessing import sequence

# IMDB dataset - 50000 samples, label 0 - negative, label - 1, positive
max_features = 10000  #uses 10000 words
maxlen = 200

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.imdb.load_data(num_words=max_features)

print("Training samples:", len(x_train))
print("Testing samples:", len(x_test))

x_train = sequence.pad_sequences(x_train, maxlen=maxlen) #Padding sequences means adding extra values (usually zeros) to make all text sequences the same length.
x_test = sequence.pad_sequences(x_test, maxlen=maxlen)

model = tf.keras.Sequential([
    tf.keras.layers.Embedding(max_features, 128, input_length=maxlen),
    tf.keras.layers.LSTM(64),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    x_train,
    y_train,
    epochs=5,
    batch_size=64,
    validation_split=0.2
)

loss, accuracy = model.evaluate(x_test, y_test)
print("\nModel Accuracy:", round(accuracy*100,2), "%")
word_index = tf.keras.datasets.imdb.get_word_index()

def encode_review(text):

    words = text.lower().split()
    encoded = []

    for word in words:
        if word in word_index and word_index[word] < max_features:
            encoded.append(word_index[word]+3)
        else:
            encoded.append(2)  # unknown word

    return sequence.pad_sequences([encoded], maxlen=maxlen)


review = input("\nEnter a movie review: ")
encoded_review = encode_review(review)
prediction = model.predict(encoded_review)
if prediction[0][0] > 0.5:
    print("\nSentiment: Positive 😊")
else:
    print("\nSentiment: Negative 😞")
```
