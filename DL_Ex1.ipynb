```python
from google.colab import files #users can upload files
import cv2 #opencv library for image processing
import matplotlib.pyplot as plt #used for plotting images

uploaded = files.upload() #used to open file chooser
filename = list(uploaded.keys())[0] #extracts the file name

img = cv2.imread(filename) #reads the image file and converts to array as (height, width, 3)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #converts the BGR into RBG


plt.imshow(img)
plt.title("Original Image")
plt.axis("off")
plt.show()

R = img[:,:,0] #takes all rows, takes all columns and extracts the red channel
G = img[:,:,1] #selects green channel
B = img[:,:,2] #selects blue channel

plt.figure(figsize=(12,4)) #creates a plotting are of 12units wide and 4units tall
plt.subplot(1,3,1); plt.imshow(R, cmap="Reds"); plt.title("Red"); plt.axis("off") #plots the red image in 1st position in 1st row, 3 columns
plt.subplot(1,3,2); plt.imshow(G, cmap="Greens"); plt.title("Green"); plt.axis("off")#plots the green image in 2nd position in 1st row, 3 columns
plt.subplot(1,3,3); plt.imshow(B, cmap="Blues"); plt.title("Blue"); plt.axis("off") #plots the blue image in 3rd position in 1st row, 3 columns
plt.show()

# ---- Binary Image using Thresholding ----
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY) #converts the rgbimage to gray
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY) #checks each pixel one by one and calculates threshold if threshold > 127 it is white else black
#threshold() returns two values like threshold used and image, we need only image so we use '_'

plt.imshow(binary, cmap="gray")
plt.title("Binary Image")
plt.axis("off")
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
if img is None:
    raise ValueError("Image not loaded.")

up_nearest = cv2.resize(img, None, fx=2, fy=2, interpolation=cv2.INTER_NEAREST) #procides image 2x larger in both height and width


down_128 = cv2.resize(img, (128, 128)) #shrinks the image size to 128*128 image
down_64  = cv2.resize(img, (64, 64))
down_32  = cv2.resize(img, (32, 32))
plt.figure(figsize=(10, 8))

plt.subplot(2, 3, 1) #plots image in 2 rows, 3 columns and in 1st position
plt.imshow(up_nearest)
plt.title("Upsampled 2 (Nearest)")
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(down_128)
plt.title("Downsampled 128128")
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(down_64)
plt.title("Downsampled 6464")
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(down_32)
plt.title("Downsampled 3232")
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

if img is None:
    raise ValueError("Image not loaded. Run Q1 code first.")

#Quantization uses Brightness, makes easier to process by simplifying the image

gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
q1  = (gray // 128) * 128  #any values < 128 is 0 and > 128 is 1 converts 0 -> 0 and 1 -> 128, provides only 2 values this is 1 bit quantization
q2  = (gray // 64) * 64 #results 0,1,2,3 and provides 0, 64, 128, 192 and produces 4 gray levels
q4  = (gray // 16) * 16 #results 0-15 and provides 16 gray levels
q8  = gray #original gray scale image
plt.figure(figsize=(10, 8))

plt.subplot(2,2,1)
plt.imshow(q1, cmap='gray')
plt.title("1-bit Quantization")
plt.axis("off")

plt.subplot(2,2,2)
plt.imshow(q2, cmap='gray')
plt.title("2-bit Quantization")
plt.axis("off")

plt.subplot(2,2,3)
plt.imshow(q4, cmap='gray')
plt.title("4-bit Quantization")
plt.axis("off")

plt.subplot(2,2,4)
plt.imshow(q8, cmap='gray')
plt.title("8-bit (Original Grayscale)")
plt.axis("off")

plt.tight_layout()
plt.show()
```
