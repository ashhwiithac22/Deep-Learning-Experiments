```python
#Laplacian detects edges in all directions
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

uploaded = files.upload() #allows the user to upload files
filename = list(uploaded.keys())[0] #stores the name of the image

img = cv2.imread(filename) #image is read in BGR color format
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #converts the image to grayscale

# Laplacian edge detection
laplacian = cv2.Laplacian(gray, cv2.CV_64F)

# Gaussian smoothing, applies gaussian blur to reduce noise
blur = cv2.GaussianBlur(gray, (5,5), 0) #opencv automatically calculates standard deviation based on kernal size(5*5)

# Laplacian of Gaussian (LoG)
log = cv2.Laplacian(blur, cv2.CV_64F)

# Display results
plt.figure(figsize=(12,8))
plt.subplot(2,2,1); plt.title("Original Grayscale")
plt.imshow(gray, cmap='gray'); plt.axis('off')

plt.subplot(2,2,2); plt.title("Laplacian")
plt.imshow(laplacian, cmap='gray'); plt.axis('off')
plt.subplot(2,2,3); plt.title("Gaussian Smoothed Image")

plt.imshow(blur, cmap='gray'); plt.axis('off')
plt.subplot(2,2,4); plt.title("LoG (Laplacian of Gaussian)")
plt.imshow(log, cmap='gray'); plt.axis('off')
plt.show()
```
