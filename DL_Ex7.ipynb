```python
'''
Edge detection is a technique used to find the boundaries of objects in an image.
3 * 3 is more sensitive to noise
5 *5 is less sensitive and produces smoother images
'''
import cv2 #opencv
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files #used to upload image from the system

uploaded = files.upload()
filename = list(uploaded.keys())[0] #stores the uploaded image name
img = cv2.imread(filename) #reads the image as color image
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #edge detection works better on grayscale so convert the color image to gray scale
ha
sobelx_3 = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)#detects edges in x direction,cv2.CV_64F is used because to store negative values and value greater than 255
sobely_3 = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3) #detects edges in y direction
sobel_3 = cv2.magnitude(sobelx_3, sobely_3)#combines both to get final edges

sobelx_5 = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=5) #uses large kernal 5 * 5
sobely_5 = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=5)
sobel_5 = cv2.magnitude(sobelx_5, sobely_5)

blur = cv2.GaussianBlur(gray, (5,5), 0) #removes noise from the image
sobelx_s = cv2.Sobel(blur, cv2.CV_64F, 1, 0, ksize=3)
sobely_s = cv2.Sobel(blur, cv2.CV_64F, 0, 1, ksize=3)
sobel_smooth = cv2.magnitude(sobelx_s, sobely_s)

plt.figure(figsize=(15,10)) #creates 15 inches height, 10 inches in width
plt.subplot(2,3,1); plt.title("Original Gray"); plt.imshow(gray, cmap='gray'); plt.axis('off') #2rows, 3columns
plt.subplot(2,3,2); plt.title("Sobel 3x3"); plt.imshow(sobel_3, cmap='gray'); plt.axis('off')
plt.subplot(2,3,3); plt.title("Sobel 5x5"); plt.imshow(sobel_5, cmap='gray'); plt.axis('off')
plt.subplot(2,3,4); plt.title("Smoothed Image"); plt.imshow(blur, cmap='gray'); plt.axis('off')
plt.subplot(2,3,5); plt.title("Sobel After Smoothing"); plt.imshow(sobel_smooth, cmap='gray'); plt.axis('off')
plt.show()
```
